```python
# 중앙 이미지 추출 및 저장 스크립트

import cv2
import os

# 설정
# ==============================
video_path = "4_stig y +.mp4"

# 출력 폴더
frames_folder = "4_stig y +/frames"
crop_folder = "4_stig y +/crops"

# 중앙 crop 좌표 (y1:y2, x1:x2)
y1, y2 = 147, 835
x1, x2 = 284, 1300

# 몇 초마다 프레임을 추출할지 (현재 1초)
frame_interval_seconds = 1
# ==============================

# 폴더 생성
os.makedirs(frames_folder, exist_ok=True)
os.makedirs(crop_folder, exist_ok=True)

# 비디오 캡처
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# interval 프레임 단위로 계산
frame_interval = int(fps * frame_interval_seconds)

frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"총 프레임 수: {frame_count}")
print(f"FPS: {fps}")
print(f"프레임 추출 간격 (frame interval): {frame_interval} 프레임마다")

frame_idx = 0
saved_frame_idx = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_idx % frame_interval == 0:
        # 원본 프레임 저장
        frame_filename = os.path.join(frames_folder, f"frame_{saved_frame_idx:04d}.png")
        cv2.imwrite(frame_filename, frame)

        # 중앙 crop
        cropped = frame[y1:y2, x1:x2]
        crop_filename = os.path.join(crop_folder, f"crop_{saved_frame_idx:04d}.png")
        cv2.imwrite(crop_filename, cropped)

        print(f"저장 완료: {frame_filename}  및  {crop_filename}")
        saved_frame_idx += 1

    frame_idx += 1

cap.release()
print("작업 완료!")
```

```python
# 파라미터 추출 및 저장 스크립트

import cv2
import os

# ==============================
# 설정
# ==============================
video_path = "4_stig y +.mp4"

# ROI crop 좌표 (stig x)
#y1, y2 = 958, 975
#x1, x2 = 1402, 1577

# ROI crop 좌표 (stig y)
y1, y2 = 1050, 1140
x1, x2 = 1600, 1755

# ROI crop 좌표 (focus)
# y1, y2 = 1001, 1140
# x1, x2 = 1648, 1755

# 프레임 추출 간격 (초)
frame_interval_seconds = 1.5

# ROI 이미지 저장 폴더
output_dir = "4_stig y +/ocr_4_stig y+"
# ==============================

# 폴더 생성
os.makedirs(output_dir, exist_ok=True)

# 비디오 캡처
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_interval = int(fps * frame_interval_seconds)

frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"총 프레임 수: {frame_count}")
print(f"FPS: {fps}")

frame_idx = 0
saved_frame_idx = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_idx % frame_interval == 0:
        roi = frame[y1:y2, x1:x2]
        
        # ROI 저장
        roi_filename = os.path.join(output_dir, f"roi_{saved_frame_idx:04d}.png")
        cv2.imwrite(roi_filename, roi)

        print(f"Saved {roi_filename}")
        saved_frame_idx += 1

    frame_idx += 1

cap.release()
print("영상 처리 완료!")
```
